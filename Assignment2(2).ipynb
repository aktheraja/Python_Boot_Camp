Count the number of occurences of the words in the file. Make a functions that gets the list of words generated in the part 1 of the problem as input and outputs a dictionary containing the number of occurences of each word.  

```python
def count_words(word_list):
    """
    The output should be a dictionary like {
        "the": 3200,
        "a": 1500,
        ...
    }
    """
    # your solution goes here
    counter = dict()
    for word in word_list:
        if word in counter:
            counter[word]=counter[word]+1# appendng words into a dictonary
        elif word not in counter:
            counter[word]=1
    return counter
    pass
print(count_words(word_list))
```

Using python's built-in `defaultdict` can help you write a shorter code. Rewrite above function using `defaultdict`. Take a look at python's documentation for more information. 

```python
from collections import defaultdict
def count_words(word_list):
    """
    Use `defaultdict` to create the same function.
    """
    d =  defaultdict(int)# using the default dictionary function
    for line in word_list:
        words = line.split()
        for word in words:
            d[word]+=1
    return d
    
    # Your solution goes here
    pass
print(count_words(word_list))
```

# The email template delimma (50 marks)

## Part 1 (5 marks)
Let's start by defining a function that gets a template and a value to fill in the template. The template is a string that contains some substrings that look like `[%some name%]`. We call them placeholders. What we want to do is to find the placeholders and replace them with appropriate values. Let's start simple and assume the only placeholder that we have is `[%name%]`. Complete the function below:

```python
def replace_name_in_template(template, name):
    # Your solution goes here
    
    placeholder="[%name%]"
    k =template.replace(placeholder,name)
    return k
pass

actual_result = replace_name_in_template("Hello [%name%]! How are you doing today [%name%]?", "John")
expected_result = 'Hello John! How are you doing today John?'
if actual_result == expected_result:
    print('Good!')
else:
    print('Wrong, "', actual_result, '" != "', expected_result)
```

## Part 2 (20 marks)

Now we're going to tweak it by supporting more placeholders, not just `[%name%]`. Complete the function below to get a `template` as a string and a `context` dictionary. The `context` provides the placeholder names and the values to be replaced. See the example for more clarification.

```python
def parse_template(template, context):
    # Your solution goes here
    
    b =context.items()
    for old,new in b:
        template = template.replace("[%"+old+"%]", new ) 
    return template
    pass


template = """Hi [%name%]!

Hope you're doing well! Can we meet on [%weekday%] [%time%] to have a brief discussion about [%subject%]?

Best,
[%my_name%]"""
context = {
    "name": "John",
    "weekday": "Thursday",
    "time": "8am",
    "subject": "some serious stuff",
    "my_name": "Lindsey",
}

actual_result = parse_template(template, context)
expected_result = """Hi John!

Hope you're doing well! Can we meet on Thursday 8am to have a brief discussion about some serious stuff?

Best,
Lindsey"""
if actual_result == expected_result:
    print('Good!')
else:
    print('Wrong, "', actual_result, '" != "', expected_result)
```

**Note that the context can have arbitrary keys. "name", "weekday", "time", ... are just some examples.**
Also, Note that just like the first part, `replace_name_in_template`, the placeholders can appear more than once. You must substitude all occurances of the placeholder with the provided value from the context dictionary.
